```python
driver = webdriver.Chrome('chromedriver')
page_num = 1
sqft_list = []
sqm_list = []
price_list = []
psf_list = []
name_list = []


location_list = []
amenities_list = []
amenities_count = []

for i in range(0,3):
    links = []
    driver.get("https://www.99.co/singapore/commercial/sale?page_num="+str(page_num))
    driver.maximize_window()
    
    lnks=driver.find_elements_by_css_selector("._2kH6B a")
    sqft = driver.find_elements_by_css_selector(".cUjn9")
    price = driver.find_elements_by_css_selector("._3XjHl")
    name = driver.find_elements_by_css_selector("._2kH6B")
    print(f"iteration:{i}")
    
    for i in range(0,len(price)):
        sqft_list.append(sqft[i].text.split("/")[0])
        sqm_list.append(sqft[i].text.split("/")[1])
        price_list.append(price[i].text.split("\n")[0])
        psf_list.append(price[i].text.split("\n")[1])
        name_list.append(name[i].text)
        links.append(lnks[i].get_attribute("href"))
        
    print(name_list[-1])
    
    for l in links:
        word = []
        driver.get(l)
        location = driver.find_elements_by_css_selector("._1p49F p")
        amenities = driver.find_elements_by_css_selector(".L9XlJ")
        for a in amenities:
            word.append(a.text)
        location_list.append(location[-1].text)
        amenities_list.append(word)
        amenities_count.append(len(word))
 
    page_num+=1
    
d = {'name': name_list, 'sqft': sqft_list, 'sqm': sqm_list,
     'psf': psf_list, 'location':location_list, 'amenities':amenities_list,'amenities_count':amenities_count,
     'price': price_list}
df = pd.DataFrame(data=d)
print(df.head(5))
print(df.info())
```
